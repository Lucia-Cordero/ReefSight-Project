```python
from PIL.ExifTags import TAGS

img_path = Image.open(img_path_dataset2)

def show_all_metadata(image_path):
    image = Image.open(image_path)
    exif_data = image._getexif()

    if not exif_data:
        print("❌ No EXIF data")
        image.close()
        return

    print("📸 ALL METADATA:")
    for tag_id, value in exif_data.items():
        tag = TAGS.get(tag_id, tag_id)
        print(f"{tag}: {value}")
    image.close()

# Check your coral photo
show_all_metadata(img_path)
```

<font color='green'> **Conclusion:** No metadata exists  </font>
