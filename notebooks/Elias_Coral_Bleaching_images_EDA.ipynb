# Get data

```python
import zipfile
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Bleached_and_Unbleached_Corals_Classification

path = '/Users/et/code/eliastopfstedt/ReefSight-Project/data/Bleached_and_Unbleached_Corals_Classification/'

# Path to the zip file
zip_file_path = path + 'archive.zip'

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(path)

# Define the paths to the extracted folders
bleached_folder = path + 'Train/Bleached'
unbleached_folder = path + 'Train/Unbleached'

# Function to load and display a random image from a folder
def load_random_image(folder_path):
    extracted_files = [f for f in os.listdir(folder_path)]
    random_file = random.choice(extracted_files)
    img_path = os.path.join(folder_path, random_file)
    img = Image.open(img_path)

    # Display the image
    plt.imshow(img)
    plt.title(random_file)
    plt.show()

# Load and display a random image from both folders
print("Displaying a random image from the 'Bleached' folder:")
load_random_image(bleached_folder)

print("Displaying a random image from the 'Unbleached' folder:")
load_random_image(unbleached_folder)
```
