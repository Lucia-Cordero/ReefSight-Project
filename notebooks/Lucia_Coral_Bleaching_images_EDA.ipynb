```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
import os
import numpy as np
from PIL import Image
```

## Dataset 1 : BHD corals

```python
# Download latest version
path1 = kagglehub.dataset_download("sonainjamil/bhd-corals")

print("Path to dataset files:", path1)
```

### Navigating the subfolders to access images

```python
os.listdir(path1)
```

```python
dataset_path1 = os.path.join(path1, "Dataset")
dataset_path1
```

```python
os.listdir(dataset_path1)
```

```python
healthy_path1 = os.path.join(dataset_path1, "Healthy")
bleached_path1 = os.path.join(dataset_path1, "Bleached")
dead_path1 = os.path.join(dataset_path1, "Dead")
```

```python
len(os.listdir(healthy_path1))
```

```python
len(os.listdir(bleached_path1))
```

```python
len(os.listdir(dead_path1))
```

<font color='green'> **Conclusion:** The are 712 (**"healthy"**), 720 (**"bleached"**) and 150 (**"dead"**) coral images </font>

### Loading one image

```python
img_name = os.listdir(healthy_path1)[11]
img_path = os.path.join(healthy_path1, img_name)
```

```python
img = Image.open(img_path)
img_array = np.array(img)
```

```python
img_array.shape
```

<font color='green'> **Conclusion:** The images are 227 x 227 pixels x  3 RGB channels  </font>

```python
plt.imshow(img_array)
```

### Checking if metadata exists

```python
exif_data = img._getexif()
print(exif_data)
```

<font color='green'> **Conclusion:** No metadata exists  </font>
